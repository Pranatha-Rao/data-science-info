### Regularization (make changes with explaination)

Since the model is fiting the tarining data but fails on the validation data it is overfitting the training data.
To counter this using regularization, two kinds of regularization
1. **Lasso**
2. **Ridge**

##### Regularization helps to solve over fitting problem which implies model performing well on training data but performing poorly on validation (test) data. Regularization solves this problem by adding a penalty term to the objective function and control the model complexity using that penalty term.

#### Ridge regularization

```python
from sklearn.linear_model import Ridge


regr = LinearRegression()
regr.fit(X_train, y_train)
rr = Ridge(alpha=0.01) # higher the alpha value, more restriction on the coefficients; low alpha > more generalization, coefficients are barely
# restricted and in this case linear and ridge regression resembles
rr.fit(X_train, y_train)
rr100 = Ridge(alpha=100) #  comparison with alpha value
rr100.fit(X_train, y_train)
train_score=regr.score(X_train, y_train)
test_score=regr.score(X_test, y_test)
Ridge_train_score = rr.score(X_train,y_train)
Ridge_test_score = rr.score(X_test, y_test)
Ridge_train_score100 = rr100.score(X_train,y_train)
Ridge_test_score100 = rr100.score(X_test, y_test)

print("linear regression train score:", train_score)
print ("linear regression test score:", test_score)
print("ridge regression train score with alpha 0.01:", Ridge_train_score)
print("ridge regression test score with alpha 0.01:", Ridge_test_score)
print("ridge regression train score with alpha 100:", Ridge_train_score100)
print("ridge regression test score with alpha 100:", Ridge_test_score100)

plt.plot(rr.coef_,alpha=0.7,linestyle='none',marker='*',markersize=5,color='red',label=r'Ridge; $\alpha = 0.01$',zorder=7) # zorder for ordering the markers
plt.plot(rr100.coef_,alpha=0.5,linestyle='none',marker='d',markersize=6,color='blue',label=r'Ridge; $\alpha = 100$') # alpha here is for transparency
```

If alpha value is 0, then it is simple linear regression.

If alpha value is 0.1, then the test score is better compared to that of linear regression.

If too high alpha value like 100 is considered, then it leads to underfitting of the data.

### Artifical Neural Network

```python
train_dataset = df_joined.sample(frac=0.8,random_state=0)
test_dataset = df_joined.drop(train_dataset.index)
```

```python
train_dataset = train_dataset.drop('Country code',axis =1)
train_dataset = train_dataset.drop('date',axis =1)
```

```python
test_dataset = test_dataset.drop('Country code',axis =1)
test_dataset = test_dataset.drop('date',axis =1)
```

```python
sns.pairplot(train_dataset, diag_kind="kde")
```

```python
train_stats = train_dataset.describe()
train_stats.pop("GDP per Capita after 2 years")
train_stats = train_stats.transpose()
train_stats
```

```python
train_labels = train_dataset.pop('GDP per Capita after 2 years')
test_labels = test_dataset.pop('GDP per Capita after 2 years')
```

```python
test_dataset = test_dataset.drop('Edu Exp', axis = 1)
train_dataset = train_dataset.drop('Edu Exp', axis = 1)
```

```python
def norm(a):
    return (a - train_stats['mean']) / train_stats['std']
```

```python
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
#from tensorflow.keras.layers import Input, Dense
```

```python

def build_model():
    model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

    optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
example_batch = normed_train_data[:9]
example_result = model.predict(example_batch)
example_result
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
earlystopper = EarlyStopping(patience=3, verbose=1)

filepath = "model.h5"

checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, 
                             save_best_only=True, mode='min')

 

callbacks_list = [earlystopper, checkpoint]
```

```python
EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=1,
  callbacks=callbacks_list)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

early_history = model.fit(normed_train_data, train_labels, 
                    epochs=EPOCHS, validation_split = 0.2, verbose=0, 
                    callbacks=[early_stop])
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
